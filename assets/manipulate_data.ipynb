```python
import pandas as pd

df_train = pd.read_csv('historical_data/german_credit_risk/wml/german_credit_data_biased_training.csv')
df_train.describe()
```

```python
import numpy as np
```

```python
#add column city -> metro, non-metro
def estimate_city(label):
    if label == 'Risk':
        if np.random.rand() <= 0.25:
            city = "Metro"
        else:
            city = "Non-Metro"
    else:
        if np.random.rand() <= 0.25:
            city = "Non-Metro"
        else:
            city = "Metro"
    return city
        
city = df_train['Risk'].apply(lambda x: estimate_city(x))
```

```python
#add column cibil -> continuous value
def estimate_cibil(label):
    if label == 'Risk':
        if np.random.rand() <= 0.2:
            cibil = 750 + np.random.randint(50,150)
        else:
            cibil = 700 + np.random.randint(0,50)
    else:
        if np.random.rand() <= 0.2:
            cibil = 700 + np.random.randint(0,50)
        else:
            cibil = 750 + np.random.randint(50,150)
    return cibil

cibil = df_train['Risk'].apply(lambda x: estimate_cibil(x))
```

```python
#add column salary
def estimate_salary(label):
    if label == 'Risk':
        if np.random.rand() <= 0.1:
            salary = 5000 + np.random.randint(0,3000)
        else:
            salary = 3000 + np.random.randint(0,2000)
    else:
        if np.random.rand() <= 0.1:
            salary = 3000 + np.random.randint(0,2000)
        else:
            salary = 5000 + np.random.randint(0,3000)
    return salary

salary = df_train['Risk'].apply(lambda x: estimate_salary(x))
```

```python
df_train['Salary'] = salary
df_train['City']= city
df_train['CIBIL'] = cibil
display(df_train)
```
